```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
n_samples = 300
n_features = 2
n_clusters_true = 3
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters_true, random_state=42)
```

### Data Before Clustering

Let's visualize the raw data points before applying K-means clustering. We expect to see some natural groupings, but the algorithm has not yet identified them.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], s=50, palette='viridis', legend=False)
plt.title('Data Before K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Performing K-Means Clustering

Now, let's apply the K-means algorithm to group the data points into the specified number of clusters (in this case, `n_clusters_true`).

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=n_clusters_true, random_state=42, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
centroids = kmeans.cluster_centers_
```

### Data After K-Means Clustering

Finally, let's visualize the data again, but this time, color-coding the points according to the clusters assigned by the K-means algorithm. The cluster centroids are also marked.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_kmeans, palette='viridis', s=50, legend='full')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids', edgecolor='black')
plt.title('Data After K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
