```python
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')

#inbuilt in the seaborn
```

```python
tips.sample(10)
```

```python
sns.scatterplot(x=tips['tip'],y=tips['total_bill'])

#Bivariate , using two numerical cols
```

```python
sns.scatterplot(x=tips['tip'],y=tips['total_bill'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])

#Multivariate , using multiple cols
```

```python
titanic = pd.read_csv('titanicDS.csv')
```

```python
titanic
```

```python
sns.barplot(x=titanic['pclass'],y=titanic['fare'])

#one col should have numerical data
```

```python
sns.barplot(x=titanic['pclass'],y=titanic['age'],hue=titanic['sex'])

#multivariate
```

```python
sns. boxplot(x=titanic['sex'],y= titanic[ 'age'])

#bivariate , x : categorical data , y : numerical data
```

```python
sns. boxplot(x=titanic['sex'],y= titanic[ 'age'], hue=titanic ['survived'])

#multivariate
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic[titanic['survived']==0]['age'],hist=False)
sns.distplot(titanic[titanic['survived']==1]['age'],hist=False)
```

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))

# heatmap needs 2D data
```

```python
sns.Clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```
