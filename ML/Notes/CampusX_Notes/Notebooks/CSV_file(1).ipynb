<h1>Opening a csv file from an URL</h1>

```python
import requests
from io import StringIO
url = "paste web link here , ex.https://raw.githubusercontent. com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv: 66.0) Gecko/20100101 Firefox/66.0"}
req = requests. get(url, headers=headers)
data = StringIO(req. text)
pd. read_csv(data)
```

converters

```python
def rename (name) :
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
pd. read_csv('IPL Matches 2008-2020.csv', converters={'team1' :rename})

#converts ma data as a dictionary nakhvano , key tarike column and value tarike function
```
