## Glass Classification

### Columns Breakdown

- **RI** → Refractive Index (how light bends through glass)  
- **Na** → Sodium content  
- **Mg** → Magnesium  
- **Al** → Aluminum  
- **Si** → Silicon (major component of glass)  
- **K** → Potassium  
- **Ca** → Calcium  
- **Ba** → Barium  
- **Fe** → Iron  

---

### Target Column (Glass Type)

- **1** → Building windows (float processed)  
- **2** → Building windows (non-float processed)  
- **3** → Vehicle windows  
- **5** → Containers  
- **6** → Tableware  
- **7** → Headlamps
